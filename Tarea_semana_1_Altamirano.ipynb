1. Cargar los datos

```python
data = pd.read_csv("https://raw.githubusercontent.com/marsgr6/EN-online/8a1cee296279c274d8f05f3905fce97c1f7636a2/data/UX_data.csv")
```

A. Descripción de las variables

**SUS (System Usability Scale):** Escala de usabilidad del sistema.

*   Valor entre 0 y 100, dado por el usuario al evaluar el sistema informático.
*   Ver: System Usability Scale (SUS)

**Time:** Tiempo en que el usuario realizó la prueba del sistema.

**Gender:** M (masculino), F (femenino).

**Age:** twenty (veinteañeros), thirty (treinta años), forty (cuarenta años).

**Kinect, NUI, Games:** texto en negrita Nivel de experiencia — none (baja), average (media), high (alta).

2. Ordenar las variables categóricas de una manera adecuada.

```python
# Definir el orden lógico para niveles de experiencia
experience_order = ['none', 'average', 'high']

# Definir el orden lógico para edades
age_order = ['twenty', 'thirty', 'forty']

# Convertir las columnas al tipo adecuado
data['ID'] = data['ID'].astype('category')
data['Gender'] = data['Gender'].astype('category')
data['Age'] = pd.Categorical(data['Age'], categories=age_order, ordered=True)
data['Kinect'] = pd.Categorical(data['Kinect'], categories=experience_order, ordered=True)
data['NUI'] = pd.Categorical(data['NUI'], categories=experience_order, ordered=True)
data['Games'] = pd.Categorical(data['Games'], categories=experience_order, ordered=True)

# Verificar los tipos de datos
data.dtypes
```

12. Diagrama de dispersión SUS vs Time agrupados por Gender:

```python
sns.lmplot(
    data=data,
    x="SUS",         # Eje X: evaluación de usabilidad
    y="Time",        # Eje Y: tiempo en realizar el ejercicio
    hue="Gender",    # Agrupado por género
    fit_reg=True,    # Incluye línea de tendencia
    palette="Set2",
    height=5,
    aspect=1.2
)

plt.title("Relación entre usabilidad (SUS) y tiempo de ejecución por género")
plt.xlabel("Puntaje SUS")
plt.ylabel("Tiempo (segundos)")
plt.show()
```

1. Relación general entre variables:

Se observa una relación inversa moderada entre SUS y Time:
a mayor puntuación de usabilidad (SUS), el tiempo de ejecución (Time) tiende a disminuir.
Esto significa que los usuarios que perciben el sistema como más fácil de usar completan las tareas en menos tiempo, lo que evidencia una buena eficiencia del diseño.

2. Diferencias por género:

**Mujeres (F):** presentan una pendiente negativa más pronunciada, pero con mayor dispersión; es decir, algunas tardan más tiempo incluso con puntajes SUS altos.
Esto indica que su percepción positiva no siempre se traduce en rapidez.

**Hombres (M): **muestran una pendiente más suave y homogénea, concentrándose en menores tiempos y mayor consistencia.
Esto refleja una relación más estable entre usabilidad percibida y desempeño.

3. Comparación con los diagramas de caja anteriores:

Sí, el resultado coincide plenamente con lo observado en los boxplots de Gender vs SUS y Gender vs Time:

* Los hombres mostraban ligeramente mejores puntajes de usabilidad (SUS) y menor tiempo promedio (Time).

* Las mujeres presentaban mayor dispersión tanto en la percepción de usabilidad como en los tiempos de uso.

4. Justificación:

El diagrama de dispersión confirma la consistencia entre percepción y desempeño:
quienes califican mejor la usabilidad son más eficientes al usar el sistema.
Las diferencias de pendiente entre géneros sugieren que la experiencia y confianza en la interfaz influyen más en las mujeres que en los hombres, aunque el patrón global es el mismo para ambos.

Existe una correlación negativa entre la usabilidad percibida (SUS) y el tiempo de ejecución (Time):
